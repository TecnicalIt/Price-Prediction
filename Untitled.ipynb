```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode = LabelEncoder()
```

```python
for column in x.columns:
    if x[column].dtype == np.number:
        continue
    x[column] =  LabelEncoder().fit_transform(x[column])
```

```python
plt.scatter(df["symboling"],df["price"],marker="*",color="red")
plt.xlabel("symboling",color="green") 
plt.ylabel("price",color="green") 
plt.title("scrap price",color=("blue"))
```

```python
from sklearn.model_selection import train_test_split
```

```python
object_ = train_test_split(x,y,test_size=.25,random_state=19)
```

```python
xtrain,xtest,ytrain,ytest = object_
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```
